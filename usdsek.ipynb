# SEK / USD

I want to explore the relationship between the Swedish krona (SEK) and the US-dollar (USD) and visualize the quarterly returns of the currency pair USD/SEK. Ie. whats the monthly return (in SEK) for holding USD.

## Data prep
To make the process a little bit easier , let's first do some data prep work:

Since all currencies are denoted in Euros we need to make some transformations to get the USD/SEK rate

```python
fx.rename(columns={'[US dollar ]': 'usd',
                   '[Swedish krona ]': 'sek',
                  'Period\\Unit:': 'date'}, # \\ = escape char for \
          inplace=True)

fx = fx.loc[:,['usd','sek','date']]


fx['date']=pd.to_datetime(fx['date'])
fx.sort_values('date', inplace=True)
fx.reset_index(drop=True, inplace=True)
```

```python
fx.usd.value_counts()
```

The '-' is in 62 of the rows for `usd`.
Lets drop these rows.

```python
fx = fx.loc[fx['usd']!='-',:]
```

```python
fx['usd'].value_counts()
```

```python
fx['sek'].value_counts()
```

The '-' is in 62 of the rows for `usd`.
Lets drop these also theses rows.

```python
fx = fx.loc[fx['sek']!='-',:]
```

```python
fx['sek'].value_counts()
```

```python
fx['usd'] = fx['usd'].astype('float')
fx['sek'] = fx['sek'].astype('float')
```

```python
fx.describe()
```

## USD/SEK currency pair 

Currency unit prices are often referenced as currency pairs and consists of a base currency and a quote currency. 

The base currency is considered the first currency within the currency pair quotation, with the second part of the quotation being the quote currency.

I want to create the currency pair 'USD/SEK' which, some what counterintuitive we get by taking `sek` / `usd`.

Since both are denoted in euros here they are already priced in the same quote currency.

```python
fx['usdsek'] = fx['sek'] / fx['usd']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(fx['date'], fx['usdsek'])
```

The plot shows the 'USD/SEK' exhange rate over time. As can be seen on the graph 'USD/SEK' rises quite sharply at three different times at around 1999-2001, 2007-2008 and 2014-2016.

The graph also shows that 'USD/SEK' seem to be trending in the longer run.
This can be seen clearer by smoothing out the prices by a moving average. 


```python
plt.plot(fx['date'], fx['usdsek'].rolling(100).mean())
```

## Quarterly returns

To make it easier to create a bar plot (less bars) we'll get the quarterly returns.

```python
import numpy as np

def quarterly(df, pricecol, datecol):
    lastdayofquarter = df.groupby(df.date.dt.to_period('Q'))['date'].max()
    df = df[df['date'].isin(lastdayofquarter)]
    prices = np.array(df[pricecol])
    ret =  prices[:-1] / prices[1:] - 1
    df = df.iloc[1:,] # Since no result for first obs.
    df['quarterly_ret'] = ret
    df['yearquarter'] = df['date'].dt.strftime('%Y-%q')
    df = df[['date','quarterly_ret']]
    return df

usdsek_q = quarterly(fx, 'usdsek', 'date')

fig, ax = plt.subplots()
ax.plot(usdsek_q['date'],usdsek_q['quarterly_ret'])
ax.tick_params(bottom=False)

plt.show()
```

Since first quarter of 2021 stopped at 8th of january we'll remove that month since we don't have the full month.

```python
usdsek_q = usdsek_q[(usdsek_q['date']>'1999-12-30') & (usdsek_q['date']<'2021-01-01')]
```

```python
# Renaming for ease of use
usdsek = usdsek_q
usdsek.columns = usdsek.columns.str.replace('quarterly_ret', 'ret')

usdsek
```

# Visualization

For the visualization i want to make I first want to separate positive and negative returns.

I'll do this by creating the new dataframes `pos` and `neg` where i'll replace negative returns with 0 for `pos` and vice versa.

```python
pos = usdsek.copy()
neg = usdsek.copy()
```

```python
pos = pos.reset_index()
pos.loc[pos['ret']<0,'ret'] = 0
pos.head()
```

```python
neg = neg.reset_index()
neg.loc[neg['ret']>0,'ret'] = 0
neg.tail()
```

```python
from matplotlib.pyplot import style
style.use('fivethirtyeight')


fig, ax = plt.subplots(figsize=(9, 5))

# Add bar for positive returns
ax.bar(range(0,len(neg)),pos.ret, color='#348ABD',alpha=0.6)

# Add bar for negative returns
# Move to bottom to make space for new xlabels between bars
ax.bar(range(0,len(neg)),neg.ret, color='#E24A33',alpha=0.6, bottom=-0.025)

# Remove grid and x/y labels
ax.grid(b=False)
ax.set_yticklabels([])
ax.set_xticklabels([])

# Create new xlabels
years = usdsek.date.dt.strftime('%Y').unique() # Get unique years in data

# Set first x-coordinate and y-coordinate
x_coord = 1
y_coord = -0.016
for year in years :
    ax.text(x_coord, y_coord, str(year),size=8, ha='center')
    x_coord += 4
    
hlinemargin = 0.046   
ax.axhline(y=-0.025, color='grey',alpha=0.5,linewidth=1, xmin=hlinemargin, xmax=1-hlinemargin)
ax.axhline(y=0.0, color='grey',alpha=0.5,linewidth=1, xmin=hlinemargin, xmax=1-hlinemargin)

# Get mean absolute return
meanabsret = np.mean(abs(usdsek['ret']))
meanabsret = str(int(round(meanabsret*100)))

# Add title
ax.text(s='USD/SEK fluctuates by +/- '+meanabsret+ '% on average',
        x=-0.7, y=.22, size=17, weight='bold')

# Add subtitle
ax.text(s='Quarterly returns for USD/SEK in percent', 
        x=-0.7, y=0.2)

# Add author and source bar at bottom
ax.text(-5, -0.2,
        'Author: Karl Karlsson' + ' '*105 + 'Data: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# Get best return
best = pos[pos['ret']==pos['ret'].max()]
beststring = '+' + str(int(round(best.ret.iloc[0]*100,0))) + '%'
ax.text(best.index[0], best.ret.iloc[0]+0.01, beststring,size=12, ha='center')
ax.bar(best.index,best.ret, color='#348ABD')


# Get worst return
worst = neg[neg['ret']==neg['ret'].min()]
worststring = '-' + str(abs(int(round(worst.ret.iloc[0]*100,0)))) + '%'
ax.text(worst.index[0], worst.ret.iloc[0]-0.045, worststring,size=12, ha='center')
ax.bar(worst.index,worst.ret, bottom=-0.025, color='#E24A33')

plt.show()
```
